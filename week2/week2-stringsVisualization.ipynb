# Generate a time series

```python
import numpy as np
radians = np.arange(0, 2*np.pi, 0.01);
```

# Plot the time series using the matplotlib library

```python
import matplotlib.pylab as plt
# plot the time series against the same time series
# https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.plot.html
# plt.plot(xvals, yvals)
plt.plot(radians, radians)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axis('tight')
plt.show()
```

# Plot the time series against the sine of the time series

```python
plt.plot(radians, np.sin(radians))
plt.xlabel('Radians')
plt.ylabel('Amplitude')
plt.show()
```

# Create strings for the x and y axis labels

```python
xlabel = 'Radians'
ylabel = 'Amplitude'
```

# Plot using the label strings

```python
plt.plot(radians, np.sin(radians))
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.show()
```

# Plot the sine wave with degrees in the x-axis instead of time

```python
# use the numpy function rad2deg to convert
plt.plot(np.rad2deg(radians), np.sin(radians))
plt.xlabel('Degrees')
plt.ylabel(ylabel)
plt.show()
```

# Import an audio file from class GitHub repo using librosa library

```python
!git clone https://github.com/jcdevaney/imc2023.git

import librosa
import librosa.display
file = 'imc2023/audioFiles/avm.wav'
sig , sr = librosa.load(file,mono=True,sr=None)
```

# Calculate the length in seconds

```python
length_s = sig.shape[0] / sr
print(length_s)
```

# Plot the time-domain representation of avm.wav

```python
plt.plot(sig)
```

# Plot with time on the x-axis

```python
time=np.arange(0,sig.shape[0])/sr
plt.plot(time,sig)
```

# Plot with amplitude on the y-axis

```python
excursion=max(abs(sig))
sigNew = sig / excursion
plt.plot(time,sigNew)
```

# Plot with labels

```python
time=np.arange(0,sig.shape[0])/sr
plt.plot(time,sigNew)
plt.xlabel('Time (seconds)')
plt.ylabel(ylabel)
```
